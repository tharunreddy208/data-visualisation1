```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')

import numpy as np
print("step1")
```

```python
fig = plt.figure()
ax = plt.axes()
ax.grid()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
print(x)
ax.plot(x, np.sin(x))
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 5, 10)
print(x)
ax.plot(x, np.sin(x))
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 5, 10)
print(x)
ax.plot(x, np.sin(x))
ax.set_title('simple plot')
ax.set_xlabel('wave length')
ax.set_ylabel('amplitude')
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 5, 10)
print(x)
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
ax.set_title('simple plot')
ax.set_xlabel('wave length')
ax.set_ylabel('amplitude')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 5, 10)
print(x)
ax.plot(x, np.sin(x), label = 'sin')
ax.plot(x, np.cos(x), label = 'cos')
ax.set_title('simple plot')
ax.set_xlabel('wave length')
ax.set_ylabel('amplitude')
ax.legend()
# ax.legend(loc=1)
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 5, 10)
print(x)
ax.plot(x, np.sin(x), label = 'sin')
ax.plot(x, np.cos(x), label = 'cos')
ax.plot(x, np.tan(x), label = 'tan')
ax.set_title('simple plot')
ax.set_xlabel('wave length')
ax.set_ylabel('amplitude')
ax.legend()
# ax.legend(loc=1)
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 5, 10)
print(x)
ax.plot(x, np.sin(x), label = 'sin', color = 'red')
ax.plot(x, np.cos(x), label = 'cos', color = 'green')
ax.set_title('simple plot')
ax.set_xlabel('wave length')
ax.set_ylabel('amplitude')
ax.legend()
```
